```python
import numpy as np
import random

# seed random numbers to make calculation
# deterministic (just a good practice)
# for use later in the code
np.random.seed(1)


# DATA GENERATION
# input dataset
numofpoints =8
X = []
X = [[x, 1] for x in range(numofpoints)]
print("\n input ordered pairs \n", X)

# output dataset
y = [2*i + 7 + (np.random.random()-0.5)*2 for i in range(numofpoints)]
print("\n output values \n", y)

# convert lists to numpyArray
X = np.array(X)
y = np.array(y).reshape(-1, 1)    #reshape() rearrages the array. 1 is the number of collumns, -1 means as many rows as neccesary.

# normalize (convert to values between 0 and 1 by dividing all values by the max value)
y_unnormalized = y
y = y / np.max(y)
print("\n input array \n", X)
print("\n output array unnormalized \n", y_unnormalized)
print("\n output array normalized \n", y)



# ALGORITHM
# syn0 = 2*np.random.random((2,1)) - 1
syn0 = [[0.1,],[-0.2,]]
syn0 = np.array(syn0)

NUM_STEPS = 1000000
lr = 0.01
print("\n  synapse 0: \n", syn0)

for iter in range(NUM_STEPS):
    # forward propagation

    # Input -> first layer
    L0 = X

    # Apply weight and bias
    L1 = np.dot(L0,syn0)

    # how much did we miss?
    L1_error = y - L1

    L1_delta = L1_error * lr

    # update weight and bias
    syn0 += np.dot(L0.T,L1_delta)


# RESULTS
denormalSynapse = syn0 * np.max(y_unnormalized)
print( "\n L1  \n ", L1)
print( "\n L1 Error :  \n", L1_error)
print( "\n L1 Delta :  \n", L1_delta)
print( "\n Output After Training (syn0):  \n", syn0)
print("\n De-Normalized Output (denormal_syn0)\n", denormalSynapse)
#best answer w/100000 epoch = (1.92, 6.81)
```
